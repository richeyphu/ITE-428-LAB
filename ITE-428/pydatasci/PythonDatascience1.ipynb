## **แบบฝึกหัดเพิ่มเติม**


---





```python
# ข้อ 1. สร้าง column ชื่อ sEngine คือผลรวมจำนวนรถจดทะเบียนขนาด 1300cc
df['sEngine'] = df['1300cc']

# ข้อ 2. สร้าง column ชื่อ mEngine คือผลรวมจำนวนรถจดทะเบียนขนาด 1600cc และ 1800ccหหหหหหหห
df['mEngine'] = df['1600cc'] + df['1800cc']

# ข้อ 3. สร้าง column ชื่อ lEngine คือผลรวมจำนวนรถจดทะเบียนขนาด 2000cc และ gt2000cc
df['lEngine'] = df['2000cc'] + df['gt2000cc']
```

```python
# ข้อ 4. แสดงข้อมูลการจดทะเบียนรถขนาดที่ใหญ่ (lEngine) ที่มากกว่า 10 คัน
df[df['lEngine']>10]
```

```python
# ข้อ 5. แสดงข้อมูลรายการที่มีการจดทะเบียนรวมของรถตั้งแต่ 1000 คันขึ้นไป
df[df['Total']>1000]
```

```python
# ข้อ 6. แสดงข้อมูลรายการที่มีการจดทะเบียนรวมของรถตั้งแต่ 100 จนถึง 500 คันขึ้นไป
# AND, OR (Python)
# Pandas AND (&) , OR (|)
df[(df['Total']>=100) & (df['Total']<=500)]
```

```python
# ข้อ 7. แสดงข้อมูลที่ไม่มีการจดทะเบียน ในบางกลุ่ม (sEngine , mEngine , lEngine)
df[(df['sEngine']==0) | (df['mEngine']==0) | (df['lEngine']==0)]
```

```python
# ข้อ 8. แสดงข้อมูลที่มีการจดทะเบียน ครบทุกกลุ่ม (sEngine , mEngine , lEngine)
df[(df['sEngine']>0) & (df['mEngine']>0) & (df['lEngine']>0)]
```

```python
# ข้อ 9. แสดงข้อมูล TOYOTA ที่มีรายการจดทะเบียน 1300cc มากกว่า 2000
df[(df['brand'].str.lower()=="toyota") & (df['1300cc']>2000)]
```

```python
# numpy   คำนวณ , เปรียบเทียบ
# IF
# np.where(เงื่อนไข, ค่าที่กำหนดให้ถ้าเงื่อนไขเป็นจริง, ค่าที่กำหนดให้่ถ้าเงื่อนไขเป็นเท็จ)
df['Status'] = np.where(df['Total']>100, "Good", "Bad")
```

```python
df
```

```python
# ข้อ 10. สร้าง column ชื่อ Focus คือ ถ้ามีการจดทะเบียน ครบทุกกลุ่ม 
# (sEngine , mEngine , lEngine) แสดงค่า valid แต่ถ้าไม่ครบทุกกลุ่มเป็น invalid
df['Focus'] = np.where((df['sEngine']>0) & (df['mEngine']>0) & (df['lEngine']>0), "valid", "invalid")
```

```python
df[(df['Focus']=="invalid")]
```

```python
# ข้อ 11. สร้าง column ชื่อ FocusSmall คือ ถ้ามีการจดทะเบียน sEngine มากที่สุด 
# โดยมากกว่า  mEngine และ  lEngine ให้แสดงค่า YES แต่ถ้าไม่ครบทุกกลุ่มเป็น NO
df['FocusSmall'] = np.where((df['sEngine']>df['mEngine']) & (df['sEngine']>df['lEngine']), "YES", "NO")
```

```python
df[(df['FocusSmall']=="YES")].shape
```

```python
# ข้อ 12. สร้าง column ชื่อ Satisfied คือ ถ้ามีการจดทะเบียน ครบทุกกลุ่ม 
# (sEngine , mEngine , lEngine)  และ มีการจดทะเบียนรวมมากกว่า 1000 คันด้วย
# ให้แสดงเป็น YES แต่ถ้าไม่ใช่ให้แสดง NO
df['Satisfied'] = np.where((df['sEngine']>0) & (df['mEngine']>0) & (df['lEngine']>0) & (df['Total']>1000), "YES", "NO")
```

```python
df[(df['Satisfied']=="YES")].shape
```

## **Aggregation**
---



```python
# sort_values(by=[],ascending=[])
see_col = ['brand','month','sEngine','mEngine','lEngine','Total']
# df[df['Total']>100][see_col].sort_values(by=['Total'], ascending=[1])
df[df['Total']>100][see_col].sort_values(by=['lEngine','mEngine'], ascending=[1,1])
```

```python
# group by
# agg --> sum , mean (avg) , min , max , count
df.describe()
```

```python
df.info()
```

```python
# agg --> sum , mean (avg) , min , max , count
see_stat = ['sum','mean','count']
df.groupby(['brand'])['Total'].agg(see_stat).sort_values(by=['sum'],ascending=[0])
```

```python
dfExport = df.groupby(['brand'])['Total'].agg(see_stat).sort_values(by=['sum'],ascending=[0])
```
