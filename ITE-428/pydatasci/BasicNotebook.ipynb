```python
import math
```

```python
def calBMI(w, h):
  # return w / (h*h)
  return w / math.pow(h, 2)
```

```python
calBMI(70, 1.7)
```

```python
print('{:.2f}'.format(calBMI(70, 1.7)))
```

```python
W = float(input("กรอกน้ำหนัก : "))
H = float(input("กรอกส่วนสูง : "))
print('BMI = {:.2f}'.format(calBMI(W, H)))
```
